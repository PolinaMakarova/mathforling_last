 Дан текст, разделенный на абзацы. Необходимо построить обратный индекс этого текста, показывающий в каком абзаце и на какой позиции хранится слово. Пользователь вводит два слова. Необходимо определить, находятся ли эти два слова в соседних абзацах. Если слова находятся в одном абзаце, необходимо найти расстояние между ними.


```python
import string
import math
```

Избавление от лишней пунктуации и больших букв (чтоб одно и то же слово не выглядело разным из-за этого

```python
def worktext(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.replace('—', '')
    return text
```

```python
with open('text.txt', 'r') as file:
    text = file.read()
text = worktext(text)
```

Получаем список по абзацам (в моем случае по одному разделителю \n) и в каждом абзаце список слов

```python
abz = text.split('\n')
for i in range(len(abz)):
    abz[i] = abz[i].split()
```

Создаем словарь: ключи - слова, значения - словарь, где ключи - индексы абзаца, значения - индексы слова в данном абзаце

```python
words = {}
```

```python
for i in range(len(abz)): # для каждого абзаца
    for j in range(len(abz[i])): # для каждого слова в абзаце
        if abz[i][j] in words.keys(): # если это слово есть в словаре
            if i in words[abz[i][j]].keys(): #если индекс абзаца есть во вложенном словаре
                words[abz[i][j]][i].append(j) # добавляем в список индексов слов данный индекс
            else: # если индекса абзаца еще нет
                words[abz[i][j]][i] = [j] #создаем список из элемента этого индекса слова
        else: # если слова еще нет в словаре
            words[abz[i][j]] = {i:[j]} # создаем значение
```

Можно посмотреть на получившийся словарь

```python
for key in words.keys():
    print(key, ': ', words[key])
```

Пользователь вводит два слова:

```python
word1 = input()
word2 = input()
```

```python
def findwords(words, word1, word2):
    if word1 not in words.keys() or word2 not in words.keys():
        print("Одного из этих слов вообще нет в тексте")
        return
    else:
        indexabz1 = words[word1].keys()
        indexabz2 = words[word2].keys()
        for ind in indexabz1:
            for ind2 in indexabz2:
                if ind == ind2:
                    print("Оба слова находятся в абзаце с индексом ", ind)
                    indword1 = words[word1][ind]
                    indword2 = words[word2][ind2]
                    between = []
                    for indw1 in indword1:
                        for indw2 in indword2:
                            between.append(int(math.fabs(indw1-indw2)))
                    betw = min(between)
                    print("Ближайшее расстояние между этими словами: ", betw)
                    return
        for ind in indexabz1:
            for ind2 in indexabz2:
                if math.fabs(ind - ind2) == 1:
                    print("Эти слова находятся в соседних абзацах: ", ind, ' и ', ind2)
                    return
    print("Эти слова не соседние никоим образом")
    return
```

```python
findwords(words, word1, word2)
```

Суть работы программы: вытаскиваем индексы абзацев, если слова есть в словаре. Сначала проходимся циклом и смотрим, есть ли они в одинаковых абзацах. если да, то считаем все близости между ними (модуль разности индексов) и выводим наименьшую. Если все это не произошло и return не сработал, то проходимся циклом еще раз, на этот раз в поиске соседних абзацев. Это нельзя было делать в одном цикле, так как могло случиться так, что одно слово есть в третьем и в четвертом абзаце, а другое только в четвертом, и он вернет то, что они соседние, когда можно было бы еще немного пройти циклом и увидеть, что в одном абзаце они тоже присутствуют. А как мне кажется, существование в одном абзаце приоритетнее, чем нахождение в соседних. А если все эти циклы прошлись и ничего не вернулось, то в конце функции выводим, что слова не соседние никак, и закрываем функцию
